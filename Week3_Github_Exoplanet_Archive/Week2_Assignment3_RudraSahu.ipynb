# Some Cool Stuff Python can do

In astronomy, observed data is typically stored in fits files, that can be plotted here using the same packages to plot data points. Here, I am going to give an example using a published image of a popular disk, that recently garnered attention for a newly discovered protoplanet in the system, AB Aur b

```python
'''
Importing the 'fits' module from the 'astropy.io' package
This module is used for reading, writing, and manipulating FITS (Flexible Image Transport System) files,
which are commonly used in astronomy for storing image and table data
'''
from astropy.io import fits
```

```python
# Loading image data from a FITS file named 'pdi_pi_collapsed.fits'
# The 'ext=1' argument specifies that the data should be read from the second HDU (Header/Data Unit)
img = fits.getdata('pdi_pi_collapsed.fits',ext=1)
```

```python
# Creating a square figure for displaying the image, with dimensions 8x8 inches
fig = plt.figure(figsize=(8,8))

# Displaying the FITS image data using imshow
# 'origin=lower' sets the [0,0] coordinate at the bottom-left
# 'vmin' and 'vmax' define the range of intensity values to be displayed
image = plt.imshow(img, origin = 'lower', vmin=-0.001, vmax=0.01)

# Adding a colorbar to indicate the surface brightness scale
# 'shrink' scales the colorbar to 82% of its default height
cbar = fig.colorbar(image,shrink=0.82)
cbar.set_label('Surface Brightness (arb. units)', rotation=90,fontsize=12)

# Setting the x-axis and y-axis limits to zoom into a specific region of interest
plt.xlim([60,140])
plt.ylim([60,140])

# Adding axis labels with LaTeX-rendered text and specified font sizes
plt.xlabel(r'$\Delta$ RA [px]',fontsize=12)
plt.ylabel(r'$\Delta$ Dec [px]',fontsize=12);
```

Python is very powerful and can read many many rows of data in a short time. In the above example, the image is a 201 x 201 matrix, so a total of 40401 data points. Now let's try reading in 20,000 rows of data with 97 different columns, and use two of those columns to create a scatter plot.

Here, we are loading positional data of 20,000 stars from a CSV table that consists of Gaia (ESA's space telescope) data for the closest stars. We are then showing all the stars as per their positional coordinates RA and Dec (Right Ascension and Declination) as follows

```python
# Importing the pandas library as 'pd' for data manipulation and analysis,
# particularly useful for handling tabular data like CSV files or DataFrames
import pandas as pd
```

```python
# Reading a CSV file containing data on the closest 20,000 stars from the Gaia Archive (DR2)
# The dataset has 97 columns of various stellar parameters
stellar=pd.read_csv('closest20kstars.csv')
stellar
```

```python
# Creating a figure for plotting with dimensions 10x10 inches
fig = plt.figure(figsize = [10,10])

# Creating a scatter plot of stellar positions
# 'ra' (Right Ascension) is plotted on the x-axis and 'dec' (Declination) on the y-axis
# 'alpha' sets the transparency of each point, 's' sets the marker size
# 'lw' sets the linewidth of marker edges, 'ec' is edge color, and 'color' is the fill color
plt.scatter(stellar['ra'], stellar['dec'], alpha=0.8, s=12, lw = 0.5, ec = 'k', color='darkorchid')

# Setting limits for the axes to ensure a uniform sky view
plt.xlim([0,362])
plt.ylim([-75,75])

# Adding axis labels with LaTeX-rendered degree symbols and appropriate font sizes
plt.xlabel(r'RA [$\degree$]', fontsize = 14)
plt.ylabel(r'Dec [$\degree$]', fontsize = 14)

# Adding a descriptive title to the plot
plt.title('Closest Stars 20,000 stars from Gaia Archive (DR2)', fontsize = 16)
```
